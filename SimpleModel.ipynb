```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as mplt
import numpy as np
import csv
import pandas as pd
```

```python
df = pd.read_csv("PandasCleaned.csv") 
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python

result = pd.DataFrame(df, columns=['precipitation']) 
result = result.to_numpy()
result = np.asarray(result).astype(np.float32)
display(result)
```

```python
from sklearn import preprocessing
del df['precipitation']
df.fillna(0, inplace=True)
display(df)
data_normalized = normalize(df)
display(data_normalized)
data = data_normalized.to_numpy()

data = np.asarray(data).astype(np.float32)

# mask = np.isnan(data)
# idx = np.where(~mask,np.arange(mask.shape[1]),0)
# np.maximum.accumulate(idx,axis=1, out=idx)
# filledData = data[np.arange(idx.shape[0])[:,None], idx]
# display(filledData)
```

```python
NeuralNetwork = keras.Sequential([
    keras.layers.Dense(5,input_shape=(41,),activation='relu')
])

NeuralNetwork.compile(
                    optimizer='adaMax', 
                    loss='mean_squared_error',
                    metrics=['mean_absolute_percentage_error']
)
#training the model
NeuralNetwork.fit(data, result, epochs=50)
```

```python
NeuralNetwork.evaluate(data, result)
```
