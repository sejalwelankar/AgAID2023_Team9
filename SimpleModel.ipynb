```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as mplt
import numpy as np
import csv
import pandas
```

```python


with open("Data.csv", 'r') as x:
    data = list(csv.reader(x, delimiter=","))
 
data = np.array(data)
display(data)

i = 0
j = 0
while(i < 851):
    while(j < 40):
        if(float(data[i][j])):
           data[i][j] = float(data[i][j])
        else:
            data[i][j] = 0
        j+=1
    i+=1
display(data)
```

```python
data.shape
```

```python

with open("Solution.csv", 'r') as x:
    solution= list(csv.reader(x, delimiter=","))
 
solution = np.array(solution)

i = 0
while(i < 851):
    if(float(solution[i][0])):
        solution[i][0] = float(solution[i][0])
    else:
        solution[i][0] = 0
    i+=1
display(solution)
```

```python
NeuralNetwork = keras.Sequential([
    keras.layers.Dense(5,input_shape=(40,),activation='relu')
])

NeuralNetwork.compile(
                    optimizer='adaMax', 
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy']
)
#training the model
NeuralNetwork.fit(data, solution, epochs=10)
```
